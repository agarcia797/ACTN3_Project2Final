# Scientific Question:
Is the ACTN3 gene sequences significantly different from the ACTN2 gene and is one more correlated with increased athletic performance? 

Background: In humans, there are two genes that encode for skeletal muscle alpha-actinins: ACTN2 and ACTN3. ACTN2 is expressed in all fibers and ACTN3 is restricted to fast (type 2) fibers. A common polymorphism in this gene is R577X, resulting in a premature stop codon, as X allele homozygotes are deficient in the ACTN3 protein which is associated with a lower fast-twitch fiber percentage but does not result in disease. The XX genotype frequency varies across different ethnic groups. In individuals of European descent, less than a third of the population have two copies of the functional R allele (the RR genotype), and just over half the population have one copy of each of the two alleles (the RX genotype). Remarkably, the remaining 18% of the healthy European population (and more than a billion people worldwide) have two copies of the nonfunctional 577X variant (the XX genotype), resulting in complete deficiency of the ACTN3 protein in their skeletal muscle. 

The data comparing the ACTN2 (https://www.ncbi.nlm.nih.gov/gene/88) and ACTN3 (https://www.ncbi.nlm.nih.gov/gene/89) gene sequences can be found from the NCBI database. The NCBI database provides many online resources for biological information and data, such as nucleotide, gene, and protein sequences.


# Scientific Hypothesis:
If the ACTN3 gene is correlated with increased endurance and sprint athletic performance, then the ACTN3 gene will be significantly different from ACTN2 gene which is expressed in non-athletes. 

Description: One aim of this project is to test for similarity between ACTN2 and ACTN3 gene sequences. Both the ACTN2 and ACTN3 gene sequences were downloaded from the NCBI database, and compared using pairwise sequence alignment, the first bioinformatics analyses performed in this notebook. 

Another goal of this project is to test whether ACTN2 or ACTN3 is correlated with enhanced sprint and endurance athletic performance. Based on the paper “ACTN3 Genotype Is Associated with Human Elite Athletic Performance” by Nan Yang, the authors use the genotypes of an elite group of athletes and a control group to analyze if ACTN3 is involved in enhanced athletic performance. The elite group of athletes were categorized into sprint or endurance groups, and were measured against the control group. Data was then obtained by their ACTN3 genotype, as wild type homozygous RR, variant heterozygous RX, or variant homozygous XX.  Considering how ACTN2 is expressed in all muscle fibers, those with ACTN3 deficiency (XX) can be used to measure ACTN2 expression against full expression of ACTN3 (RR). The data from this study was used to measure any statistical difference between how the ACTN3 and ACTN2 gene correlates to enhanced sprint and endurance performance by p-values, the second bioinformatics analysis done in this notebook. A csv file (http://localhost:8888/edit/Desktop/BIMM%20143/ACTN3table.csv) was created utilizing the data from the above paper, and analyzed for statistical significance between control and athletic groups. 

Based on this data, a multiple bar graph was created to show whether ACTN3 or ACTN2 was more potent than the other utilizing genotype frequencies. To accurately measure both endurance and sprint athletic performance, a subset of the elite athletes (http://localhost:8888/edit/Desktop/BIMM%20143/ACTN3plot.csv) was created and categorized as "Sprint Olympians" or "Endurance Olympians", representing those elite athletes who competed at the Olympic level.

```python
# Import packages needed
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
from Bio import SeqIO
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Part 2: Load in the data and perform Bioinformatics Analyses (Pairwise Sequence Alignment & P-value analysis)

Both ACTN3 and ACTN2 are separate FASTA files, obtained from the NCBI database, that contain the ACTN3 and ACTN2 gene sequences which are found in Homo sapiens on chromosome 11. These gene sequences are compared using pairwise sequence alignment by considering the entire length of the two gene sequences and matching them to obtain the best alignment. A score is provided based on their similarity. 

A csv file titled “ACTN3table.csv” was manually created and read in using pandas. Each group is compared based on the ACTN3 genotype, and a p-value is calculated to measure statistical significance between presence and absence of the ACTN3 gene on athletic performance.




```python
# Read in ACTN3 gene sequence from NCBI database
# Define a global variable ACTN3
# prints out ACTN3 gene sequence
ACTN3 = SeqIO.read("ACTN3 sequence.fasta", "fasta")
print(ACTN3)
```

```python
# Read in ACTN2 gene sequence from NCBI database
# Define a global variable ACTN2
# prints out ACTN2 gene sequence
ACTN2 = SeqIO.read("ACTN2 sequence.fasta", "fasta")
print(ACTN2)
```

```python
# Define two sequences to be aligned for pairwise sequence alignment
# The score of alignment is calculated by the sum of substitution and gap scores
# Define a global variable 'alignments'
# Obtain a similarity score of 210
alignments = pairwise2.align.globalxx(ACTN3, ACTN2)
for a in alignments:
    print(format_alignment(*a))
```

```python
# Perform statistical analysis using p-values to check for any difference between the ACTN3 gene and athletic performance
# Table represents genotype frequencies in all control, endurance, and sprint groups for both male and female (along with a combined total)
# Code check to see if 'ACTN3table.csv' file read in correctly. Uncheck when complete
#table = pd.read_csv("ACTN3table.csv")
#table
# Sample size of each group is defined in p-value calculation
# Define a global variable ACTN3males
# Read in csv file and sort into male group
ACTN3males = pd.read_csv("ACTN3table.csv", nrows=3, index_col=0)
ACTN3males
```

```python
# Use tool 'stats' from Biopython function SciPy to measure statistical difference and obtain p-value
# reads in two samples to test if they have the same statistical properties
# returns a statistical score depending on similarity and a p-value
#Define local variable 'MaleControlsRR'
#Define local variable 'MaleControlsXX'
#There is a significant difference between the male control group with the ACTN3 gene and those without (ACTN3 deficient)
MaleControlsRR = stats.norm.rvs(loc = 30,scale = 10,size = 40)
MaleControlsXX = stats.norm.rvs(loc = 16,scale = 10,size = 21)
print(stats.ttest_ind(MaleControlsRR, MaleControlsXX))

#Define a local variable 'MaleSprintRR'
#Define a local variable 'MaleSprintXX'
# There is a significant difference between the male sprint group with the ACTN3 gene and those without (ACTN3 deficient)
MaleSprintRR = stats.norm.rvs(loc = 53,scale = 10,size = 38)
MaleSprintXX = stats.norm.rvs(loc = 8,scale = 10,size = 6)
print(stats.ttest_ind(MaleSprintRR, MaleSprintXX))

#Define a local variable 'MaleEnduranceRR'
##Define a local variable 'MaleEnduranceXX'
# There is a significant difference between the male endurance group with the ACTN3 gene and those without (ACTN3 deficient)
#Highest p-value in male set
MaleEnduranceRR = stats.norm.rvs(loc = 28,scale = 10,size = 34)
MaleEnduranceXX = stats.norm.rvs(loc = 20,scale = 10,size = 25)
print(stats.ttest_ind(MaleEnduranceRR, MaleEnduranceXX))
```

```python
# Define a global variable ACTN3females
# Read in csv file and sort into female group
ACTN3females = pd.read_csv("ACTN3table.csv", skiprows=5, nrows=3, index_col= 0)
ACTN3females
```

```python
#Define a local variable 'FemaleControlsRR'
#Define a local variable 'FemaleControlsXX'
# There is a significant difference between the female control group with the ACTN3 gene and those without (ACTN3 deficient)
FemaleControlsRR = stats.norm.rvs(loc = 30,scale = 10,size = 88)
FemaleControlsXX = stats.norm.rvs(loc = 20,scale = 10,size = 57)
print(stats.ttest_ind(FemaleControlsRR, FemaleControlsXX))

#Define a local variable 'FemaleSprintRR'
#Define a local variable 'FemaleSprintXX'
# No defined p-value, due to female sprint group without ACTN3 gene having 0 sample size. Can assume that a significant difference exists
FemaleSprintRR = stats.norm.rvs(loc = 43,scale = 10,size = 15)
FemaleSprintXX = stats.norm.rvs(loc = 0,scale = 10,size = 0)
print(stats.ttest_ind(FemaleSprintRR, FemaleSprintXX))

#Define a local variable 'FemaleEnduranceRR'
#Define a local variable 'FemaleEnduranceXX'
#No significant difference between female endurance group with the ACTN3 gene and those without (ACTN3 deficient)
#Highest p-value in female set
FemaleEnduranceRR = stats.norm.rvs(loc = 36,scale = 10,size = 26)
FemaleEnduranceXX = stats.norm.rvs(loc = 29,scale = 10,size = 21)
print(stats.ttest_ind(FemaleEnduranceRR, FemaleEnduranceXX))
```

```python
# Define ACTN3total
# Read in csv file and sort into total group
ACTN3total = pd.read_csv("ACTN3table.csv", skiprows=10, nrows=3, index_col= 0)
ACTN3total
```

```python
#Define a local variable 'TotalControlsRR'
#Define a local variable 'TotalControlsXX'
# There is a significant difference between the total control group with the ACTN3 gene and those without (ACTN3 deficient)
TotalControlsRR = stats.norm.rvs(loc = 30,scale = 10,size = 130)
TotalControlsXX = stats.norm.rvs(loc = 18,scale = 10,size = 80)
print(stats.ttest_ind(TotalControlsRR, TotalControlsXX))

#Define a local variable 'TotalSprintRR'
#Define a local variable 'TotalSprintXX'
# there is a significant difference between total sprint group with the ACTN3 gene and those without (ACTN3 deficient)
TotalSprintRR = stats.norm.rvs(loc = 50,scale = 10,size = 53)
TotalSprintXX = stats.norm.rvs(loc = 6,scale = 10,size = 6)
print(stats.ttest_ind(TotalSprintRR, TotalSprintXX))

#Define a local variable 'TotalEnduranceRR'
#Define a local variable 'TotalEnduranceXX'
# There is a significant difference between total endurance group with the ACTN3 gene and those without (ACTN3 deficient)
#Highest p value in total set
TotalEnduranceRR = stats.norm.rvs(loc = 31,scale = 10,size = 60)
TotalEnduranceXX = stats.norm.rvs(loc = 24,scale = 10,size = 46)
print(stats.ttest_ind(TotalEnduranceRR, TotalEnduranceXX))
```

# Part 3: Plotting the Results

Here we are using matplotlib to create the multiple bar graph shown below. We create a multiple bar graph to show the presence of ACTN3 across endurance/sprint athletic and control groups. In this case, the presence of ACTN3 will show us whether it is involved in enhanced athletic performance. The x labels are created using the "ACTN3plotx.csv"  file so it can be read in using pandas. A legend is created using the columns in the csv file, which displays the presence of ACTN3 (RR) in blue and the absence of ACTN3 (XX), thus greater expression of ACTN2, in orange. 

A multiple bar plot is a visualization technique to compare different entities. A bar plot represents a category of data with rectangular bars, composed of lengths and heights that is proportional to the values which they represent, as demonstrated on the y-axis. Here, the graph is used to represent the genotype frequencies in both the athletic and control groups, with different categories of groups on the x-axis and the genotype frequencies in percentages on the y-axis. 


```python
# A csv file was created for the x label groups
#Define a global variable RR
#check that csv file was correctly read in
# code check
col_names = ['RR', 'XX']
df = pd.read_csv("ACTN3plot.csv")
RR = df['RR']
#print(RR)
```

```python
# Define a global variable XX
#code check
XX= df['XX']
#print(XX)
```

```python
# Subset data to measure olympians

df = pd.read_csv("ACTN3plotx.csv")
#code check to see if x-axis labels were read
#df
#define a global variable 'x'
x= df['x axis']

# setting the width of the bars
# define local variable 'w'
w=0.4

#setting the number of bars in the graph and how to be positioned next to each other
# define local variables 'bar1' and 'bar2'
bar1 = np.arange(len(x)) #[0,1,2,3,4,5,6]
bar2 = [i+w for i in bar1]

#setting bar to data types
#bar1 represents the 'RR' genotype frequencies
#bar2 represents the 'XX' genotype frequencies
plt.bar(bar1,RR,w,label= "RR")
plt.bar(bar2,XX,w,label= "XX")

#setting x and y labels
plt.xlabel("Athletic & Control Groups")
plt.ylabel("Genotype Frequencies (%)")
#setting a title for the graph
plt.title("Genotype Frequencies in Athletic vs. Control Groups")
#labeling and positioning x labels
plt.xticks(bar1+w/2, x, rotation = 75)
# inputting a legend categorizing 'RR' and 'XX' genotypes
plt.legend()

plt.show()
```

# Part 4: Analysis of the results

The plot above shows the presence of both ACTN3 (wildtype RR genotype) and ACTN2 (variant XX genotype) among various groups. Based on the multiple bar graph, it is shown that the ACTN3 gene is significantly present among power/sprint groups, and ACTN3 expression decreases among control and endurance athletic groups. Among both control and endurance athletic groups, ACTN2 is more dominant, with its highest expression in endurance olympians. 

Although genotype frequencies represent significant differences between those with ACTN3 and those without (p-value<0.05) in male and total groups, the results of the bar graph also align with the p-values calculated considering how the highest obtained p-value from each set (male, female, and totals) were the endurance group when comparing differences in genotype expression between ACTN3 (RR) and ACTN2 (XX). It is significant to note that athletic performance may differ between males and females, as the female endurance group had no significant difference in genotype frequencies between those with ACTN3 vs. ACTN3 deficiency. Males had a significant difference among all control, endurance, and sprint groups, implying that ACTN3 may affect enhanced athletic performance differently in males in comparison to females. 

Based on the pairwise sequence alignment between ACTN3 and ACTN2 gene sequences, a match score of 210 was given. The score of alignment is calculated by the sum of substitution and gap scores, and a high positive score represents many sums (similarities) and little gaps (differences). Thus, we can conclude that ACTN3 and ACTN2 gene sequences are very similar to each other, although some differences may exist which may contribute to greater expression of ACTN3 among the sprint group in the multiple bar graph. Therefore, my hypothesis was not correct as ACTN3 and ACTN2 gene sequences are not significantly different from each other, and ACTN3 (RR) is greatly expressed among power/sprint groups which confers a greater advantage upon sprint performance compared to endurance performance.
